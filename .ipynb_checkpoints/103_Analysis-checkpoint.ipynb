This is the fourth and final code in a series designed to thoroughly analyze the impact of the PMGSY road construction program in India on rural development, using nighttime light data as an indicator of economic activity. This document focuses on conducting a comprehensive econometric analysis.

**1. Data Used in this Code:**
   - **Final Dataset**: A comprehensive dataset that integrates Road Data, Nighttime Light Data, Terrain & Elevation Data, and Shrid Data, obtained through data extraction, merging, and wrangling as performed in Code-100, Code-101, and Code-102.

**2. Objectives of this Code:**
   - Perform descriptive and econometric analysis to assess the impact of the PMGSY road construction program on rural development using nighttime light data.
   - Perform PSM
   - Perform DiD
   - Perform Triple DiD
   - Perform RKD
   - Perform Robustness Check
     
**3. Results:**
   - Analyze the impact of the PMGSY road construction program on rural development using the comprehensive zonal dataset.
   - Evaluate significant increases or decreases in nighttime light following road construction.
   - Examine the influence of terrain and zonal differences.
   - Conduct robustness checks to validate the results.

This document details the steps and code used, ensuring that the analysis is transparent and reproducible.

```python
# Visualizations

# Filter the dataframe for rows where has_road is True
has_road_df = df[df['has_road'] == 1]

# Get the unique shrids and their first completion year
unique_shrid_first_completion_year = has_road_df.drop_duplicates(subset=['shrid'])[['shrid', 'first_completion_year']]

# Calculate the frequency of each unique first completion year
unique_shrid_freq = unique_shrid_first_completion_year['first_completion_year'].value_counts().sort_index()

# Calculate the total number of unique shrids that have received roads
total_unique_shrids = unique_shrid_first_completion_year['shrid'].nunique()

# Calculate the mode year of first completion
mode_year = unique_shrid_freq.idxmax()

print(f'Total number of unique shrids that have received roads: {total_unique_shrids}')
print(f'Mode year of first completion: {mode_year}')

# Fit a normal distribution to the data
(mu, sigma) = norm.fit(unique_shrid_freq.values)

# Plotting the frequency of unique first completion year
plt.figure(figsize=(14, 7))
bars = plt.bar(unique_shrid_freq.index, unique_shrid_freq.values, color='#76c7c0', edgecolor='black', linewidth=1.2, alpha=0.7)

# Add the number of shrids on each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom', ha='center', fontsize=10, fontweight='bold')

# Plot the fitted curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p * max(unique_shrid_freq.values) * (xmax-xmin)/len(unique_shrid_freq.index), 'b--', linewidth=2)

# Add summary box
textstr = '\n'.join((
    f'MEAN: {mu:.2f}',
    f'SD: {sigma:.2f}',
    f'Total Unique shrids: {total_unique_shrids}',
    f'Mode Year: {mode_year}'
))

# Create a textbox
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.text(0.80, 0.99, textstr, transform=plt.gca().transAxes, fontsize=12,
        verticalalignment='top', bbox=props)

plt.xlabel('First Road Completion Year', fontsize=14)
plt.ylabel('Frequency of Shrid', fontsize=14)
plt.title('Frequency of First Road Completion Year of the Shrid for Overall Country (INDIA)', fontsize=16, fontweight='bold')
plt.xticks(np.arange(min(unique_shrid_freq.index), max(unique_shrid_freq.index)+1, 1.0), rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# # Add gridlines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# # Add a background color
plt.gca().set_facecolor('#f7f7f7')

# # Remove top and right spines
sns.despine()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Zonal Visualization

# Filter the dataframe for rows where has_road is True
has_road_df = df[df['has_road'] == 1]

# Get the unique shrids and their first completion year
unique_shrid_first_completion_year = has_road_df.drop_duplicates(subset=['shrid'])[['shrid', 'first_completion_year', 'zone']]

# Create a 3x2 grid for plotting
fig, axes = plt.subplots(3, 2, figsize=(18, 18))
fig.suptitle('Frequency of First Road Completion Year of the Shrid by Zone', fontsize=24, fontweight='bold')

# List of zones
zones = unique_shrid_first_completion_year['zone'].unique()

# Plot for each zone
for i, zone in enumerate(zones):
    ax = axes[i//2, i%2]
    zone_data = unique_shrid_first_completion_year[unique_shrid_first_completion_year['zone'] == zone]
    
    # Calculate the frequency of each unique first completion year
    unique_shrid_freq = zone_data['first_completion_year'].value_counts().sort_index()
    
    # Calculate the total number of unique shrids that have received roads
    total_unique_shrids = zone_data['shrid'].nunique()
    
    # Calculate the mode year of first completion
    mode_year = unique_shrid_freq.idxmax()
    
    # Fit a normal distribution to the data
    (mu, sigma) = norm.fit(unique_shrid_freq.values)
    
    # Plotting the frequency of unique first completion year
    bars = ax.bar(unique_shrid_freq.index, unique_shrid_freq.values, color='#008080', edgecolor='black', linewidth=1.2, alpha=0.7)
    
    # Add the number of shrids on each bar
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom', ha='center', fontsize=10, fontweight='bold')
    
    # Add summary box
    textstr = '\n'.join((
        f'MEAN: {mu:.2f}',
        f'SD: {sigma:.2f}',
        f'Total Unique shrids: {total_unique_shrids}',
        f'Mode Year: {mode_year}'
    ))
    
    # Create a textbox and place it outside the bar plot
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    ax.annotate(textstr, xy=(0.80, 0.99), xycoords='axes fraction', fontsize=11,
                verticalalignment='top', bbox=props)
    
    ax.set_xlabel('First Road Completion Year', fontsize=14, fontweight='light')
    ax.set_ylabel('Frequency of Shrids', fontsize=14, fontweight='light')
    ax.set_title(f'Zone: {zone}', fontsize=16, fontweight='bold')
    
    # Set x-ticks and labels
    xticks = np.arange(min(unique_shrid_freq.index), max(unique_shrid_freq.index) + 1)
    ax.set_xticks(xticks)
    ax.set_xticklabels(xticks, rotation=45, fontsize=12)
    
    # Set y-ticks font size
    for label in ax.get_yticklabels():
        label.set_fontsize(12)
    
    # Add gridlines
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Add a background color
    ax.set_facecolor('#f7f7f7')
    
    # Remove top and right spines
    sns.despine(ax=ax)

# Remove any unused subplots
for j in range(i+1, len(axes.flat)):
    fig.delaxes(axes.flat[j])

# Adjusting the layout to prevent overlap
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
#Staggered Visualization

# Filter the dataframe for rows where has_road is True
has_road_df = df[df['has_road'] == 1]

# Get the unique shrids and their first completion year
unique_shrid_first_completion_year = has_road_df.drop_duplicates(subset=['shrid'])[['shrid', 'first_completion_year']]

# Calculate the frequency of each unique first completion year
unique_shrid_freq = unique_shrid_first_completion_year['first_completion_year'].value_counts().sort_index()

# Calculate the total number of unique shrids that have received roads
total_unique_shrids = unique_shrid_first_completion_year['shrid'].nunique()

# Calculate the mode year of first completion
mode_year = unique_shrid_freq.idxmax()

print(f'Total number of unique shrids that have received roads: {total_unique_shrids}')
print(f'Mode year of first completion: {mode_year}')

# Calculate years relative to mode year
df['years_relative'] = df['year'] - mode_year

# Separate data for has_road and no_road
has_road_data = df[df['has_road'] == 1]
no_road_data = df[df['has_road'] == 0]

# Calculate the median night light value by years relative to mode year
has_road_median_night_light = has_road_data.groupby('years_relative')['night_light_value'].median()
no_road_median_night_light = no_road_data.groupby('years_relative')['night_light_value'].median()

# Create the plot
plt.figure(figsize=(14, 7))

# Plot the median night light values as scatter plots
plt.scatter(has_road_median_night_light.index, has_road_median_night_light.values, marker='o', label='Has Road')
plt.scatter(no_road_median_night_light.index, no_road_median_night_light.values, marker='x', label='No Road')
plt.axvline(x=0, color='red', linestyle='--')

# Best fit line for negative values
neg_indices = has_road_median_night_light.index < 0
pos_indices = has_road_median_night_light.index > 0

# Has Road - Negative
slope, intercept, _, _, _ = linregress(has_road_median_night_light.index[neg_indices], has_road_median_night_light.values[neg_indices])
plt.plot(has_road_median_night_light.index[neg_indices], intercept + slope * has_road_median_night_light.index[neg_indices], 'b--', linewidth=2, label='Best Fit Has Road (Negative)')

# Has Road - Positive
slope, intercept, _, _, _ = linregress(has_road_median_night_light.index[pos_indices], has_road_median_night_light.values[pos_indices])
plt.plot(has_road_median_night_light.index[pos_indices], intercept + slope * has_road_median_night_light.index[pos_indices], 'b-', linewidth=2, label='Best Fit Has Road (Positive)')

# No Road - Negative
neg_indices = no_road_median_night_light.index < 0
pos_indices = no_road_median_night_light.index > 0
slope, intercept, _, _, _ = linregress(no_road_median_night_light.index[neg_indices], no_road_median_night_light.values[neg_indices])
plt.plot(no_road_median_night_light.index[neg_indices], intercept + slope * no_road_median_night_light.index[neg_indices], 'g--', linewidth=2, label='Best Fit No Road (Negative)')

# No Road - Positive
slope, intercept, _, _, _ = linregress(no_road_median_night_light.index[pos_indices], no_road_median_night_light.values[pos_indices])
plt.plot(no_road_median_night_light.index[pos_indices], intercept + slope * no_road_median_night_light.index[pos_indices], 'g-', linewidth=2, label='Best Fit No Road (Positive)')

# Set the x-axis limits
plt.xlim(-15, 15)

# Add summary box
textstr = '\n'.join((
    f'Total Unique shrids: {total_unique_shrids}',
    f'Mode Year: {mode_year}'
))

# Create a textbox
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.text(0.80, 0.10, textstr, transform=plt.gca().transAxes, fontsize=12,
         verticalalignment='top', bbox=props)

# Set titles and labels
plt.xlabel('Years Relative to First Completion Year (MODE)', fontsize=14)
plt.ylabel('Median Night Light Value', fontsize=14)
plt.title('Night Light Intensity Before and After Road Completion by Overall Relative to MODE Year of Country for both Treated & Control Shrids', fontsize=16, fontweight='bold')

# Set x-ticks and labels
xticks = np.arange(-15, 16, 1)
plt.xticks(xticks, rotation=45, fontsize=12)

# Set y-ticks font size
plt.yticks(fontsize=12)

# Add gridlines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add a background color
plt.gca().set_facecolor('#f7f7f7')

# Remove top and right spines
sns.despine()

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

## Econometric Analysis

```python

# Function for data preprocessing
def preprocess_data(data):
    mode_treatment_year = data.loc[data['has_road'], 'first_completion_year'].mode()[0]
    data['treatment_year'] = mode_treatment_year
    data['years_from_treatment'] = data['year'] - data['treatment_year']
    data['post_treatment'] = data['years_from_treatment'] >= 0
    data['treated'] = data['has_road']
    data['DiD'] = data['post_treatment'] * data['treated']
    
    tri_bins = [0, 3.25, 5.68, 7.97, float('inf')]
    tri_labels = ["Level Terrain", "Slightly Rugged Terrain", "Moderately Rugged Terrain", "Extremely Rugged Terrain"]
    data['tri_mean_tertile'] = pd.cut(data['tri_mean'], bins=tri_bins, labels=tri_labels)
    
    # Handling NaNs before converting to int
    tri_interaction = pd.cut(data['tri_mean'], bins=tri_bins, labels=False)
    tri_interaction = tri_interaction.fillna(-1).astype(int)  # Fill NaNs with -1 and then convert to int
    data['tri_interaction'] = tri_interaction
    
    data['Triple_DiD'] = data['DiD'] * data['tri_interaction']
    data['time_trend'] = data['year'] - data['year'].min()
    
    return data

def propensity_score_matching(data):
    X = data[['area_sq_km', 'elevation_mean']]
    y = data['treated']
    X_scaled = StandardScaler().fit_transform(X)
    ps_model = LogisticRegression(max_iter=1000).fit(X_scaled, y)
    data['pscore'] = ps_model.predict_proba(X_scaled)[:, 1]

    treated = data[data['treated'] == 1]
    control = data[data['treated'] == 0]
    knn = NearestNeighbors(n_neighbors=1).fit(control[['pscore']])
    distances, indices = knn.kneighbors(treated[['pscore']])
    matched_control = control.iloc[indices.flatten()]

    matched_data = pd.concat([treated, matched_control])
    
    # Standardize the relevant variables, including night_light_value, in matched data
    scaler = StandardScaler()
    matched_data[['tri_mean', 'area_sq_km', 'elevation_mean', 'time_trend', 'night_light_value']] = scaler.fit_transform(
        matched_data[['tri_mean', 'area_sq_km', 'elevation_mean', 'time_trend', 'night_light_value']]
    )
    
    return matched_data, data

def calculate_smd(original_data, matched_data):
    def smd(x, y):
        return (x.mean() - y.mean()) / np.sqrt((x.var() + y.var()) / 2)

    covariates = ['tri_mean', 'area_sq_km', 'elevation_mean']
    smd_before = {}
    smd_after = {}

    for covariate in covariates:
        smd_before[covariate] = smd(original_data.loc[original_data['treated'] == 1, covariate],
                                    original_data.loc[original_data['treated'] == 0, covariate])
        smd_after[covariate] = smd(matched_data.loc[matched_data['treated'] == 1, covariate],
                                   matched_data.loc[matched_data['treated'] == 0, covariate])

    smd_df = pd.DataFrame({'SMD Before': smd_before, 'SMD After': smd_after})
    print("Standardized Mean Differences (SMD) Before and After Matching:")
    print(smd_df)

# Function to fit the DiD model
def fit_did_model(data):
    model = smf.ols('night_light_value ~ post_treatment * treated + time_trend', data=data).fit(cov_type='HC3')
    return model

# Function to fit the Triple DiD model
def fit_didid_model(data):
    model = smf.ols('night_light_value ~ post_treatment * treated * tri_interaction + time_trend', data=data).fit(cov_type='HC3')
    return model

# Function to interpret the DiD model results
def interpret_did_results(model):
    print("DiD Model Results:")
    print("Post-Treatment Period:")
    print(f"  Coefficient (post_treatment[T.True]): {model.params['post_treatment[T.True]']:.4f}")
    print(f"  Interpretation: {'Increase' if model.params['post_treatment[T.True]'] > 0 else 'Decrease'} in nighttime light intensity after treatment period.")
    
    print("\nImpact of Treatment:")
    print(f"  Coefficient (treated[T.True]): {model.params['treated[T.True]']:.4f}")
    print(f"  Interpretation: {'Higher' if model.params['treated[T.True]'] > 0 else 'Lower'} baseline nighttime light intensity for treated villages compared to control villages.")
    print(f"  Coefficient (post_treatment[T.True]:treated[T.True]): {model.params['post_treatment[T.True]:treated[T.True]']:.4f}")
    print(f"  Interpretation: {'Increase' if model.params['post_treatment[T.True]:treated[T.True]'] > 0 else 'Decrease'} in nighttime light intensity for treated villages post-treatment compared to control villages post-treatment.")
    
    print("\nTime Trend:")
    print(f"  Coefficient (time_trend): {model.params['time_trend']:.4f}")
    print(f"  Interpretation: {'Increase' if model.params['time_trend'] > 0 else 'Decrease'} in nighttime light intensity over time.")

# Function to interpret the Triple DiD model results
def interpret_didid_results(model):
    print("Triple DiD Model Results:")
    print("Post-Treatment Period:")
    print(f"  Coefficient (post_treatment[T.True]): {model.params['post_treatment[T.True]']:.4f}")
    print(f"  Interpretation: {'Increase' if model.params['post_treatment[T.True]'] > 0 else 'Decrease'} in nighttime light intensity after treatment period.")
    
    print("\nImpact of Treatment:")
    print(f"  Coefficient (treated[T.True]): {model.params['treated[T.True]']:.4f}")
    print(f"  Interpretation: {'Higher' if model.params['treated[T.True]'] > 0 else 'Lower'} baseline nighttime light intensity for treated villages compared to control villages.")
    print(f"  Coefficient (post_treatment[T.True]:treated[T.True]): {model.params['post_treatment[T.True]:treated[T.True]']:.4f}")
    print(f"  Interpretation: {'Increase' if model.params['post_treatment[T.True]:treated[T.True]'] > 0 else 'Decrease'} in nighttime light intensity for treated villages post-treatment compared to control villages post-treatment.")
    
    print("\nTerrain Ruggedness:")
    print(f"  Coefficient (tri_interaction): {model.params['tri_interaction']:.4f}")
    print(f"  Interpretation: {'Higher' if model.params['tri_interaction'] > 0 else 'Lower'} nighttime light intensity in more rugged terrain.")
    print(f"  Coefficient (post_treatment[T.True]:tri_interaction): {model.params['post_treatment[T.True]:tri_interaction']:.4f}")
    print(f"  Interpretation: {'Increase' if model.params['post_treatment[T.True]:tri_interaction'] > 0 else 'Decrease'} in nighttime light intensity post-treatment in more rugged terrain.")
    print(f"  Coefficient (treated[T.True]:tri_interaction): {model.params['treated[T.True]:tri_interaction']:.4f}")
    print(f"  Interpretation: {'Higher' if model.params['treated[T.True]:tri_interaction'] > 0 else 'Lower'} nighttime light intensity for treated villages in more rugged terrain.")
    print(f"  Coefficient (post_treatment[T.True]:treated[T.True]:tri_interaction): {model.params['post_treatment[T.True]:treated[T.True]:tri_interaction']:.4f}")
    print(f"  Interpretation: {'Higher' if model.params['post_treatment[T.True]:treated[T.True]:tri_interaction'] > 0 else 'Lower'} increase in nighttime light intensity for treated villages post-treatment in more rugged terrain.")

# Function to analyze pre-treatment trend
def analyze_pre_treatment_trend(data):
    pre_treatment_data = data[data['post_treatment'] == False]
    
    # Fit a model to check the trend in the pre-treatment period
    pre_trend_model = smf.ols('night_light_value ~ time_trend * treated', data=pre_treatment_data).fit(cov_type='HC3')
    
    print("\nPre-Treatment Trend Analysis Results:")
    print(pre_trend_model.summary())
    
    # Interpretation of Parallel Trends Assumption for DiD
    if abs(pre_trend_model.params['time_trend:treated[T.True]']) < 0.05:
        print("The coefficient for the interaction between time_trend and treated is close to zero, indicating that the parallel trends assumption for DiD is likely valid.")
    else:
        print("The coefficient for the interaction between time_trend and treated is not close to zero, indicating that the parallel trends assumption for DiD may not be valid.")
    
    # Visualizing the pre-treatment trend
    plt.figure(figsize=(12, 6))
    sns.lineplot(x='time_trend', y='night_light_value', hue='treated', data=pre_treatment_data, markers=True, style='treated')
    
    # Fit regression lines
    treated_subset = pre_treatment_data[pre_treatment_data['treated'] == 1]
    control_subset = pre_treatment_data[pre_treatment_data['treated'] == 0]
    
    slope_treated, intercept_treated, _, _, _ = stats.linregress(treated_subset['time_trend'], treated_subset['night_light_value'])
    plt.plot(treated_subset['time_trend'], intercept_treated + slope_treated * treated_subset['time_trend'], 'orange', label='Treated (regression)')
    
    slope_control, intercept_control, _, _, _ = stats.linregress(control_subset['time_trend'], control_subset['night_light_value'])
    plt.plot(control_subset['time_trend'], intercept_control + slope_control * control_subset['time_trend'], 'blue', label='Control (regression)')
    
    plt.title('Pre-Treatment Trend in Nighttime Light Intensity')
    plt.xlabel('Time Trend (Years)')
    plt.ylabel('Nighttime Light Intensity')
    plt.legend(title='Treated')
    plt.show()

    return pre_trend_model

# Function to analyze pre-treatment trend for Triple DiD with subplots for each terrain and treatment condition
def analyze_pre_treatment_trend_triple_did(data):
    pre_treatment_data = data[data['post_treatment'] == False]
    
    # Fit a model to check the trend in the pre-treatment period for Triple DiD
    pre_trend_model = smf.ols('night_light_value ~ time_trend * treated * tri_interaction', data=pre_treatment_data).fit(cov_type='HC3')
    
    print("\nPre-Treatment Trend Analysis for Triple DiD Results:")
    print(pre_trend_model.summary())
    
    # Interpretation of Parallel Trends Assumption for Triple DiD
    if abs(pre_trend_model.params['time_trend:treated[T.True]']) < 0.05 and abs(pre_trend_model.params['time_trend:treated[T.True]:tri_interaction']) < 0.05:
        print("The coefficients for the interactions involving time_trend, treated, and tri_interaction are close to zero, indicating that the parallel trends assumption for Triple DiD is likely valid.")
    else:
        print("The coefficients for the interactions involving time_trend, treated, and tri_interaction are not close to zero, indicating that the parallel trends assumption for Triple DiD may not be valid.")
    
    # Define terrain labels
    terrain_labels = ["Level Terrain", "Slightly Rugged Terrain", "Moderately Rugged Terrain", "Extremely Rugged Terrain"]
    
    # Create a 2x2 grid of subplots
    fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharex=True, sharey=True)
    axes = axes.flatten()
    
    for i, label in enumerate(terrain_labels):
        ax = axes[i]
        subset = pre_treatment_data[pre_treatment_data['tri_mean_tertile'] == label]
        sns.lineplot(x='time_trend', y='night_light_value', hue='treated', style='treated', data=subset, markers=True, ax=ax)
        
        # Fit regression lines
        treated_subset = subset[subset['treated'] == 1]
        control_subset = subset[subset['treated'] == 0]
        
        slope_treated, intercept_treated, _, _, _ = stats.linregress(treated_subset['time_trend'], treated_subset['night_light_value'])
        ax.plot(treated_subset['time_trend'], intercept_treated + slope_treated * treated_subset['time_trend'], 'orange', label='Treated (regression)')
        
        slope_control, intercept_control, _, _, _ = stats.linregress(control_subset['time_trend'], control_subset['night_light_value'])
        ax.plot(control_subset['time_trend'], intercept_control + slope_control * control_subset['time_trend'], 'blue', label='Control (regression)')
        
        ax.set_title(label)
        ax.set_xlabel('Time Trend (Years)')
        ax.set_ylabel('Nighttime Light Intensity')
        ax.legend(title='Treated')

    plt.tight_layout()
    plt.suptitle('Pre-Treatment Trend in Nighttime Light Intensity for Triple DiD', y=1.02)
    plt.show()

    return pre_trend_model

# Main function to run the analysis
def perform_analysis(data, analysis_level='entire', target_zone=None, target_state=None, target_tri=None):
    # Subset data based on the analysis level
    if analysis_level == 'entire':
        data_subset = data.copy()
    elif analysis_level == 'zone':
        if target_zone is None:
            raise ValueError("Please specify the target_zone for zone-level analysis.")
        data_subset = data[data['zone'] == target_zone].copy()
    elif analysis_level == 'state':
        if target_state is None:
            raise ValueError("Please specify the target_state for state-level analysis.")
        data_subset = data[data['state_name'] == target_state].copy()
    elif analysis_level == 'tri':
        if target_tri is None:
            raise ValueError("Please specify the target_tri for terrain-level analysis.")
        data_subset = data[data['tri_mean_tertile'] == target_tri].copy()
    else:
        raise ValueError("Invalid analysis_level. Choose from 'entire', 'zone', 'state', or 'tri'.")
    
    print(f"Data Subset Count: {len(data_subset)}")
    
    # Preprocess data
    data_subset = preprocess_data(data_subset)
    print(f"Data Preprocessed: Count after preprocessing = {len(data_subset)}")
    
    # Apply PSM
    matched_data, original_data = propensity_score_matching(data_subset)
    print(f"PSM Done: Matched Data Count = {len(matched_data)}")
    
    # Calculate SMD
    calculate_smd(original_data, matched_data)
    
    # Fit the models
    did_model = fit_did_model(matched_data)
    print("\nDiD Model Results:")
    print(did_model.summary())
    
    didid_model = fit_didid_model(matched_data)
    print("\nTriple DiD Model Results:")
    print(didid_model.summary())

    # Interpret results
    print("\nInterpreting Results for DiD Model:")
    interpret_did_results(did_model)

    print("\nInterpreting Results for Triple DiD Model:")
    interpret_didid_results(didid_model)
    
    # Analyze pre-treatment trend
    analyze_pre_treatment_trend(matched_data)
    analyze_pre_treatment_trend_triple_did(matched_data)
```

```python
# For Zonal Analysis


# Function for data preprocessing
def preprocess_data(data):
    mode_treatment_year = data.loc[data['has_road'], 'first_completion_year'].mode()[0]
    data['treatment_year'] = mode_treatment_year
    data['years_from_treatment'] = data['year'] - data['treatment_year']
    data['post_treatment'] = data['years_from_treatment'] >= 0
    data['treated'] = data['has_road']
    data['DiD'] = data['post_treatment'] * data['treated']
    
    tri_bins = [0, 3.25, 5.68, 7.97, float('inf')]
    tri_labels = ["Level Terrain", "Slightly Rugged Terrain", "Moderately Rugged Terrain", "Extremely Rugged Terrain"]
    data['tri_mean_tertile'] = pd.cut(data['tri_mean'], bins=tri_bins, labels=tri_labels)
    
    # Handling NaNs before converting to int
    tri_interaction = pd.cut(data['tri_mean'], bins=tri_bins, labels=False)
    tri_interaction = tri_interaction.fillna(-1).astype(int)  # Fill NaNs with -1 and then convert to int
    data['tri_interaction'] = tri_interaction
    
    data['Triple_DiD'] = data['DiD'] * data['tri_interaction']
    data['time_trend'] = data['year'] - data['year'].min()
    
    return data

def propensity_score_matching(data):
    X = data[['area_sq_km', 'elevation_mean']]
    y = data['treated']
    X_scaled = StandardScaler().fit_transform(X)
    ps_model = LogisticRegression(max_iter=1000).fit(X_scaled, y)
    data['pscore'] = ps_model.predict_proba(X_scaled)[:, 1]

    treated = data[data['treated'] == 1]
    control = data[data['treated'] == 0]
    knn = NearestNeighbors(n_neighbors=1).fit(control[['pscore']])
    distances, indices = knn.kneighbors(treated[['pscore']])
    matched_control = control.iloc[indices.flatten()]

    matched_data = pd.concat([treated, matched_control])
    
    # Standardize the relevant variables, including night_light_value, in matched data
    scaler = StandardScaler()
    matched_data[['tri_mean', 'area_sq_km', 'elevation_mean', 'time_trend', 'night_light_value']] = scaler.fit_transform(
        matched_data[['tri_mean', 'area_sq_km', 'elevation_mean', 'time_trend', 'night_light_value']]
    )
    
    return matched_data, data

def calculate_smd(original_data, matched_data):
    def smd(x, y):
        return (x.mean() - y.mean()) / np.sqrt((x.var() + y.var()) / 2)

    covariates = ['tri_mean', 'area_sq_km', 'elevation_mean']
    smd_before = {}
    smd_after = {}

    for covariate in covariates:
        smd_before[covariate] = smd(original_data.loc[original_data['treated'] == 1, covariate],
                                    original_data.loc[original_data['treated'] == 0, covariate])
        smd_after[covariate] = smd(matched_data.loc[matched_data['treated'] == 1, covariate],
                                   matched_data.loc[matched_data['treated'] == 0, covariate])

    smd_df = pd.DataFrame({'SMD Before': smd_before, 'SMD After': smd_after})
    print("Standardized Mean Differences (SMD) Before and After Matching:")
    print(smd_df)

# Function to fit the DiD model
def fit_did_model(data):
    model = smf.ols('night_light_value ~ post_treatment * treated + time_trend', data=data).fit(cov_type='HC3')
    return model

# Function to fit the Triple DiD model
def fit_didid_model(data):
    model = smf.ols('night_light_value ~ post_treatment * treated * tri_interaction + time_trend', data=data).fit(cov_type='HC3')
    return model

# Function to interpret the DiD model results
def interpret_did_results(model):
    print("DiD Model Results:")
    print("Post-Treatment Period:")
    print(f"  Coefficient (post_treatment[T.True]): {model.params['post_treatment[T.True]']:.4f}")
    print(f"  Interpretation: {'Increase' if model.params['post_treatment[T.True]'] > 0 else 'Decrease'} in nighttime light intensity after treatment period.")
    
    print("\nImpact of Treatment:")
    print(f"  Coefficient (treated[T.True]): {model.params['treated[T.True]']:.4f}")
    print(f"  Interpretation: {'Higher' if model.params['treated[T.True]'] > 0 else 'Lower'} baseline nighttime light intensity for treated villages compared to control villages.")
    print(f"  Coefficient (post_treatment[T.True]:treated[T.True]): {model.params['post_treatment[T.True]:treated[T.True]']:.4f}")
    print(f"  Interpretation: {'Increase' if model.params['post_treatment[T.True]:treated[T.True]'] > 0 else 'Decrease'} in nighttime light intensity for treated villages post-treatment compared to control villages post-treatment.")
    
    print("\nTime Trend:")
    print(f"  Coefficient (time_trend): {model.params['time_trend']:.4f}")
    print(f"  Interpretation: {'Increase' if model.params['time_trend'] > 0 else 'Decrease'} in nighttime light intensity over time.")

# Function to interpret the Triple DiD model results
def interpret_didid_results(model):
    print("Triple DiD Model Results:")
    print("Post-Treatment Period:")
    print(f"  Coefficient (post_treatment[T.True]): {model.params['post_treatment[T.True]']:.4f}")
    print(f"  Interpretation: {'Increase' if model.params['post_treatment[T.True]'] > 0 else 'Decrease'} in nighttime light intensity after treatment period.")
    
    print("\nImpact of Treatment:")
    print(f"  Coefficient (treated[T.True]): {model.params['treated[T.True]']:.4f}")
    print(f"  Interpretation: {'Higher' if model.params['treated[T.True]'] > 0 else 'Lower'} baseline nighttime light intensity for treated villages compared to control villages.")
    print(f"  Coefficient (post_treatment[T.True]:treated[T.True]): {model.params['post_treatment[T.True]:treated[T.True]']:.4f}")
    print(f"  Interpretation: {'Increase' if model.params['post_treatment[T.True]:treated[T.True]'] > 0 else 'Decrease'} in nighttime light intensity for treated villages post-treatment compared to control villages post-treatment.")
    
    print("\nTerrain Ruggedness:")
    print(f"  Coefficient (tri_interaction): {model.params['tri_interaction']:.4f}")
    print(f"  Interpretation: {'Higher' if model.params['tri_interaction'] > 0 else 'Lower'} nighttime light intensity in more rugged terrain.")
    print(f"  Coefficient (post_treatment[T.True]:tri_interaction): {model.params['post_treatment[T.True]:tri_interaction']:.4f}")
    print(f"  Interpretation: {'Increase' if model.params['post_treatment[T.True]:tri_interaction'] > 0 else 'Decrease'} in nighttime light intensity post-treatment in more rugged terrain.")
    print(f"  Coefficient (treated[T.True]:tri_interaction): {model.params['treated[T.True]:tri_interaction']:.4f}")
    print(f"  Interpretation: {'Higher' if model.params['treated[T.True]:tri_interaction'] > 0 else 'Lower'} nighttime light intensity for treated villages in more rugged terrain.")
    print(f"  Coefficient (post_treatment[T.True]:treated[T.True]:tri_interaction): {model.params['post_treatment[T.True]:treated[T.True]:tri_interaction']:.4f}")
    print(f"  Interpretation: {'Higher' if model.params['post_treatment[T.True]:treated[T.True]:tri_interaction'] > 0 else 'Lower'} increase in nighttime light intensity for treated villages post-treatment in more rugged terrain.")


# Main function to run the analysis
def perform_analysis(data, analysis_level='entire', target_zone=None, target_state=None, target_tri=None):
    # Subset data based on the analysis level
    if analysis_level == 'entire':
        data_subset = data.copy()
    elif analysis_level == 'zone':
        if target_zone is None:
            raise ValueError("Please specify the target_zone for zone-level analysis.")
        data_subset = data[data['zone'] == target_zone].copy()
    elif analysis_level == 'state':
        if target_state is None:
            raise ValueError("Please specify the target_state for state-level analysis.")
        data_subset = data[data['state_name'] == target_state].copy()
    elif analysis_level == 'tri':
        if target_tri is None:
            raise ValueError("Please specify the target_tri for terrain-level analysis.")
        data_subset = data[data['tri_mean_tertile'] == target_tri].copy()
    else:
        raise ValueError("Invalid analysis_level. Choose from 'entire', 'zone', 'state', or 'tri'.")
    
    print(f"Data Subset Count: {len(data_subset)}")
    
    # Preprocess data
    data_subset = preprocess_data(data_subset)
    print(f"Data Preprocessed: Count after preprocessing = {len(data_subset)}")
    
    # Apply PSM
    matched_data, original_data = propensity_score_matching(data_subset)
    print(f"PSM Done: Matched Data Count = {len(matched_data)}")
    
    # Calculate SMD
    calculate_smd(original_data, matched_data)
    
    # Fit the models
    did_model = fit_did_model(matched_data)
    print("\nDiD Model Results:")
    print(did_model.summary())
    
    didid_model = fit_didid_model(matched_data)
    print("\nTriple DiD Model Results:")
    print(didid_model.summary())

    # Interpret results
    print("\nInterpreting Results for DiD Model:")
    interpret_did_results(did_model)

    print("\nInterpreting Results for Triple DiD Model:")
    interpret_didid_results(didid_model)
    

```

```python
#Robustness Check


# Function for data preprocessing
def preprocess_data(data, treatment_year):
    data['treatment_year'] = treatment_year
    data['years_from_treatment'] = data['year'] - data['treatment_year']
    data['post_treatment'] = data['years_from_treatment'] >= 0
    data['treated'] = data['has_road']
    data['DiD'] = data['post_treatment'] * data['treated']
    
    tri_bins = [0, 3.25, 5.68, 7.97, float('inf')]
    tri_labels = ["Level Terrain", "Slightly Rugged Terrain", "Moderately Rugged Terrain", "Extremely Rugged Terrain"]
    data['tri_mean_tertile'] = pd.cut(data['tri_mean'], bins=tri_bins, labels=tri_labels)
    
    # Handling NaNs before converting to int
    tri_interaction = pd.cut(data['tri_mean'], bins=tri_bins, labels=False)
    tri_interaction = tri_interaction.fillna(-1).astype(int)  # Fill NaNs with -1 and then convert to int
    data['tri_interaction'] = tri_interaction
    
    data['Triple_DiD'] = data['DiD'] * data['tri_interaction']
    data['time_trend'] = data['year'] - data['year'].min()
    
    return data

def propensity_score_matching(data):
    X = data[['area_sq_km', 'elevation_mean']]
    y = data['treated']
    X_scaled = StandardScaler().fit_transform(X)
    ps_model = LogisticRegression(max_iter=1000).fit(X_scaled, y)
    data['pscore'] = ps_model.predict_proba(X_scaled)[:, 1]

    treated = data[data['treated'] == 1]
    control = data[data['treated'] == 0]
    knn = NearestNeighbors(n_neighbors=1).fit(control[['pscore']])
    distances, indices = knn.kneighbors(treated[['pscore']])
    matched_control = control.iloc[indices.flatten()]

    matched_data = pd.concat([treated, matched_control])
    
    # Standardize the relevant variables, including night_light_value, in matched data
    scaler = StandardScaler()
    matched_data[['tri_mean', 'area_sq_km', 'elevation_mean', 'time_trend', 'night_light_value']] = scaler.fit_transform(
        matched_data[['tri_mean', 'area_sq_km', 'elevation_mean', 'time_trend', 'night_light_value']]
    )
    
    return matched_data, data

# Function to fit the DiD model
def fit_did_model(data):
    model = smf.ols('night_light_value ~ post_treatment * treated + time_trend', data=data).fit(cov_type='HC3')
    return model

# Function to fit the Triple DiD model
def fit_didid_model(data):
    model = smf.ols('night_light_value ~ post_treatment * treated * tri_interaction + time_trend', data=data).fit(cov_type='HC3')
    return model

# Function to perform robustness check
def robustness_check(data, analysis_level='overall', target_zone=None, treatment_years=None):
    if treatment_years is None:
        treatment_years = [1996, 2002, 2007, 2009, 2011, 2015, 2018]
    
    results = []

    for year in treatment_years:
        print(f"\nRunning analysis for treatment year: {year}")

        # Preprocess data with the specified treatment year
        data_subset = preprocess_data(data.copy(), year)
        
        if analysis_level == 'zone':
            if target_zone is None:
                raise ValueError("Please specify the target_zone for zone-level analysis.")
            data_subset = data_subset[data_subset['zone'] == target_zone].copy()
        
        # Apply PSM
        matched_data, original_data = propensity_score_matching(data_subset)
        
        # Fit the models
        did_model = fit_did_model(matched_data)
        didid_model = fit_didid_model(matched_data)
        
        # Extract the relevant coefficients
        did_coef = did_model.params['post_treatment[T.True]:treated[T.True]']
        didid_coef = didid_model.params['post_treatment[T.True]:treated[T.True]:tri_interaction']
        
        # Collect results
        results.append({
            'Year': year,
            'DiD Coefficient (post_treatment[T.True]:treated[T.True])': did_coef,
            'Triple DiD Coefficient (post_treatment[T.True]:treated[T.True]:tri_interaction)': didid_coef,
            'DiD Time Trend (time_trend)': did_model.params['time_trend'],
            'Triple DiD Time Trend (time_trend)': didid_model.params['time_trend']
        })

    # Convert results to DataFrame
    results_df = pd.DataFrame(results)
    print(results_df)
    return results_df
```

```python
# Regression Kink Design Analysis

# Function for data preprocessing specific for RKD
def preprocess_data_rkd(data):
    data['assignment_var'] = data['year'] - data['first_completion_year']
    data['post_treatment'] = (data['assignment_var'] >= 0).astype(int)
    data['time_trend'] = data['year'] - data['year'].min()
    data['treated'] = data['has_road']
    return data

# Function to fit the RKD model
def fit_rkd_model(data):
    # RKD Model Specification
    rkd_model = smf.ols('night_light_value ~ assignment_var * post_treatment + time_trend', data=data).fit(cov_type='HC3')
    return rkd_model

# Function to interpret the RKD model results
def interpret_rkd_results(model):
    print("RKD Model Results:")
    print(f"Assignment Variable (Slope before treatment): {model.params['assignment_var']:.4f}")
    print(f"Post-Treatment Interaction (Change in slope after treatment): {model.params['assignment_var:post_treatment']:.4f}")
    print(f"Post-Treatment Indicator (Immediate level change): {model.params['post_treatment']:.4f}")
    print(f"Time Trend: {model.params['time_trend']:.4f}")

    print("\nDetailed Summary:")
    print(model.summary())

# Function to plot RKD results and fitted values
def plot_rkd_results(data, model):
    sns.set(style="whitegrid")
    
    fig, axs = plt.subplots(1, 2, figsize=(24, 8))
    # Scatter plot with regression lines
    ax = axs[0]
    # Regression lines
    assignment_var = np.linspace(data['assignment_var'].min(), data['assignment_var'].max(), 100)
    pre_treatment = assignment_var[assignment_var < 0]
    post_treatment = assignment_var[assignment_var >= 0]
    
    # Predicted values
    pre_pred = model.params['Intercept'] + model.params['assignment_var'] * pre_treatment + model.params['time_trend'] * (data['time_trend'].min())
    post_pred = model.params['Intercept'] + model.params['post_treatment'] + model.params['assignment_var'] * post_treatment + model.params['assignment_var:post_treatment'] * post_treatment + model.params['time_trend'] * (data['time_trend'].min())
    
    ax.plot(pre_treatment, pre_pred, color='blue', label='Pre-Treatment Trend', linewidth=2)
    ax.plot(post_treatment, post_pred, color='red', label='Post-Treatment Trend', linewidth=2)
    
    ax.axvline(x=0, color='grey', linestyle='--', label='Treatment Threshold', linewidth=1.5)
    ax.set_title('Regression Kink Design (RKD) Results', fontsize=16)
    ax.set_xlabel('Assignment Variable (Years from Treatment)', fontsize=14)
    ax.set_ylabel('Nighttime Light Intensity', fontsize=14)
    ax.legend(fontsize=12)
    
    # Fitted values plot
    data['fitted_values'] = model.fittedvalues
    ax = axs[1]
    sns.lineplot(x='assignment_var', y='fitted_values', data=data, hue='post_treatment', palette=['blue', 'red'], ax=ax, linewidth=2)
    ax.axvline(x=0, color='grey', linestyle='--', label='Treatment Threshold', linewidth=1.5)
    ax.set_title('Fitted Values from RKD Model', fontsize=16)
    ax.set_xlabel('Assignment Variable (Years from Treatment)', fontsize=14)
    ax.set_ylabel('Fitted Nighttime Light Intensity', fontsize=14)
    ax.legend(title='Post Treatment', labels=['Pre-Treatment', 'Post-Treatment'], fontsize=12)
    
    plt.tight_layout()
    plt.show()

# Function to perform the RKD analysis
def perform_rkd_analysis(data):
    # Preprocess data for RKD
    data = preprocess_data_rkd(data)
    print(f"Data Preprocessed for RKD: Count = {len(data)}")
    
    # Fit and interpret RKD model
    rkd_model = fit_rkd_model(data[data['treated'] == 1])  # Focus on treated group only
    interpret_rkd_results(rkd_model)
    
    # Plot RKD results
    plot_rkd_results(data[data['treated'] == 1], rkd_model)


```
